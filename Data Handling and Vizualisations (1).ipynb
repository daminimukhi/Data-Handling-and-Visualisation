```python
import matplotlib.pyplot as plt

# Define points
x1, y1 = 3, 5
x2, y2 = 7, 2

# Calculate the incline and stop of the line passing through the points
incline = (y2 - y1) / (x2 - x1)
stop = y1 - incline * x1

# Define x values for the line
x_values = [x1, x2]

# Calculate corresponding y values for the line
y_values = [incline * x + stop for x in x_values]

# Plot the points and line
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, linestyle='-', color='black', linewidth=2, label='Line through points')
plt.scatter([x1, x2], [y1, y2], color='black', s=150, marker='o', label='Points')

# Set axis limits and labels
plt.xlim(0, 10)
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16, fontweight='bold')
plt.ylabel('Y', fontsize=16, fontweight='bold')
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')

# Add grid with subtle lines
plt.grid(True, linestyle=':', alpha=0.6)

# Add legend with larger font size and bold text
plt.legend(fontsize=14, loc='upper right')

# Add a title with larger font size and bold text
plt.title('line segment plot', fontsize=18, fontweight='bold')

# Adjust aspect ratio for better visualization
plt.gca().set_aspect('equal', adjustable='box')

# Show plot
plt.show()
```
